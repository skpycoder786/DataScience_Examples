# 1) Distplots :-

Distplot stands for distribution plot, it takes as
input an array and plots a curve corresponding to
the distribution of points in the array.
The seaborn.distplot() function is used to plot the
distplot. The distplot represents the univariate
distribution of data i.e. data distribution of a
variable against the density distribution.

The seaborn.distplot() function accepts the data
variable as an argument and returns the plot with
the density distribution.


```python
# .distplot Fn. of seaborn module
import matplotlib.pyplot as plt
import seaborn as sb
sb.distplot([0,1,2,3,4,5])  # difference is 1 between each pair
plt.title('First DistPlot')
plt.show()

sb.distplot([0,2,5,9,14,20])  # difference is different between each pair
plt.title('Second DistPlot')
plt.show()
```

```python
# .random.randn() Fn. of numpy module
# rest modules already attached
import numpy as np
data=np.random.randn(100)  # will generate 100 random numbers
print(data)
sb.distplot(data)
plt.title('Third DistPlot')
plt.show()
```

# 2) Line Plots :-

The line plot is one of the most basic plot in
seaborn library. Seaborn Line Plots depict the
relationship between continuous as well as
categorical values in a continuous data point
format.

Syntax for creating line plot is >
seaborn.lineplot(x, y, data)

x: Data variable for the x-axis
,y: Data variable for the y-axis
,data: The object pointing to the entire data set or
data values

```python
# .lineplot() Fn. of seaborn module
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd

x= [10,5,15,20,7,8]
y= [100,120,150,95,210,240]
data_plot = pd.DataFrame({"Product_no":x, "Cost":y})
print(data_plot)
sb.lineplot(x='Product_no', y='Cost',data=data_plot)
plt.title('First Line Plot')
plt.show()
```

```python
# all modules already attached
x=[1,3,5,7]
y=[10,20,5,30]
sb.lineplot(x,y,color='red')
plt.title('Second Line Plot')
```

# 4) Marginal plot :-

A marginal plot allows to study the relationship
between 2 numeric variables. The central chart display
their correlation. It is usually a scatterplot,
a hexbin plot, a 2D histogram or a 2D density plot. The
marginal charts, usually at the top and at the right,
show the distribution of the 2 variables
using histogram or density plot.

The seaborn library
provides a joint plot function that is really handy to
make this type of graphic. jointplot() allows you to
basically match up two distplots for bivariate
data. With your choice of what kind parameter to
compare with: “scatter” “reg” “resid” “kde” “hex”.

```python
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np

df=sb.load_dataset('iris')    # 'iris' is also a Built in data set.
print(df)
```

```python
# .jointplot() Fn. of seaborn module
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np

df=sb.load_dataset('iris')    # 'iris' is also a Built in data set.

# custom the inside plot: optins are: “scatter” “reg” “resid” “kde” “hex”.
# case-1
sb.jointplot(x=df["petal_length"], y=df["petal_width"], kind="scatter")
```

```python
# case-2
sb.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind="reg")
```

```python
# case-3
sb.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind="resid")
```

```python
# case-4
sb.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind="kde")
```

```python
# case-5
sb.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind="hex")
```

# 5) Stripplot :-

The stripplot will draw a scatterplot where one
variable is categorical.

A strip plot can be drawn on its
own, but it is also a good complement to a box or violin
plot in cases where you want to show all observations
along with some representation of the underlying
distribution

```python
# .stripplot() Fn. of seaborn module
# all modules already attached
tips=sb.load_dataset('tips')
sb.stripplot(x="day", y="total_bill", data=tips) # here 'x' is categorical variable
print("In stripplot, points are overlapped.")
```

# 6) Swarmplot :-

The swarmplot is similar to stripplot(), but the points
are adjusted (only along the categorical axis) so that
they don’t overlap.

This gives a better representation
of the distribution of values, although it does not scale
as well to large numbers of observations.

```python
# .swarmplot() Fn. of seaborn module
# all modules already attached
tips=sb.load_dataset('tips')
sb.swarmplot(x="day", y="total_bill", data=tips) # here 'x' is categorical variable
print("Whereas in swarmplot, points are not overlapped.")
```

## boxplots and violinplots are used to shown the distribution of categorical data.

A box plot (or box-and-whisker plot)
shows the distribution of quantitative data in a way that
facilitates comparisons between variables or across levels of
a categorical variable. The box shows the quartiles of the
dataset.

A violin plot plays a similar role as a box and
whisker plot. It shows the distribution of quantitative data
across several levels of one (or more) categorical variables
such that those distributions can be compared.

Unlike a box
plot, in which all of the plot components correspond to
actual datapoints, the violin plot features a kernel density
estimation of the underlying distribution.


# 7) Boxplot :-

```python
# .boxplot() Fn. of seaborn module
# all modules already attached
tips=sb.load_dataset('tips')
sb.boxplot(x="day", y="total_bill", data=tips) # here 'x' is categorical variable
print("In boxplot, all of the plot components correspond to actual datapoints. ")
```

# 8) Violinplot :-

```python
# .violinplot() Fn. of seaborn module
# all modules already attached
tips=sb.load_dataset('tips')
sb.violinplot(x="day", y="total_bill", data=tips) # here 'x' is categorical variable
print("Whereas the violin plot features a kernel density estimation of the underlying distribution. ")
```
