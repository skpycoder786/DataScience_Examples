# 1) 3D Line Plot :-

The most basic three-dimensional plot is a 3D line
plot created from sets of (x, y, z) triples. This can be
created using the ax.plot3D function.

First, we have to install matplotlib to import the
mplot3d toolkit. Mplot3d is a toolkit which will
help for matplotlib to generate 3 axes on the
graph. A three-dimensional axes can be created by
passing the keyword projection='3d' to any of the
normal axes creation routines.

```python
# 'mplot3d' is Submodule of matplotlib and ax.plot3D(), ax.set_title() are Fn.s of it.
from mpl_toolkits import mplot3d    
import matplotlib.pyplot as plt    # .axes() Fn.
import numpy as np    # .linspace(), .sin(), .cos() Fn.s

ax = plt.axes(projection='3d')    # Becuase of 'mplot3d' this line is enable. 
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)

ax.plot3D(x, y, z, 'gray')    # module 'mplot3d' 
ax.set_title('First 3D line plot')
plt.show()
```

#  2) 3D Scatter Plot :-

3D scatter plots are used to plot data points on
three axes in the attempt to show the relationship
between three variables.

To create a 3D Scatter plot, Matplotlibs mplot3d
toolkit is used to enable three dimensional
plotting. Generally 3D scatter plot is created by
using ax.scatter3D() the function of the matplotlib
library which accepts a data sets of X, Y and Z to
create the plot while the rest of the attributes of
the function are the same as that of two
dimensional scatter plot

```python
from mpl_toolkits import mplot3d    # ax.scatter3D() Fn.
import numpy as np
import matplotlib.pyplot as plt

ax = plt.axes(projection='3d')  # Compulsary line for 3D plotation
z = [5,10,15,20,25]
x = [1,2,3,4,5]
y = [2,4,6,8,10]

ax.scatter3D(x, y, z)    # no color change, blue is fix.
ax.set_title('First 3D Scatter plot')
plt.show()
```

```python
# All modules imported already
ax = plt.axes(projection='3d')    # Compulsary line for 3D plotation
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)

ax.scatter3D(x, y, z)
ax.set_title('Second 3D Scatter plot')
plt.show()
```

# 3) 3D Contour Plot :-

A contour plot is appropriate if you want to see
how value Z changes as a function of two inputs X
and Y, such that Z = f(X,Y). A contour line of a
function of two variables is a curve along which the
function has a constant value.

The ax.contour3D() function creates threedimensional contour plot. It requires all the input
data to be in the form of two-dimensional regular
grids, with the Z-data evaluated at each point. The
3d plots are enabled by importing the mplot3d
toolkit.

```python
from mpl_toolkits import mplot3d    # ax.contour3D(), ax.set_title(), ax.set_xlabel(), ax.set_ylabel(), ax.set_zlabel() Fn.s
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X ** 2 + Y ** 2))  # it is little bit different from circle's equation

ax = plt.axes(projection='3d')    # Compulsary line for 3D plotation
ax.contour3D(X, Y, Z, 40, cmap='binary')
# Here 40 is showing frequeny, if it will increase then difference between lines wil decrease.
# cmap is'color map' but there are few colors are permitted. u can see them by raising an error in cmap.
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('First 3D contour')
plt.show()
```

```python
# All modules imported already
xlist = np.linspace(-3.0, 3.0, 100)    # previous ex. in 2D graphs.
ylist = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)    # it is the Circle's equation

ax = plt.axes(projection='3d')    # Compulsary line for 3D plotation
ax.contour3D(X, Y, Z, 200, cmap='Accent')  # u can use more different values of cmap
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z-axis')
ax.set_title('3D contour')
plt.show()
```

# 4) 3D Surface Plot :-

A Surface Plot is a representation of threedimensional dataset. It describes a functional
relationship between two independent variables X
and Z and a designated dependent variable Y,
rather than showing the individual data points.

It is
a companion plot of the contour plot. It is similar
to the wireframe plot, but each face of the
wireframe is a filled polygon. This helps to create
the topology of the surface which is being
visualized.
Surface plots are created by using the
ax.plot_surface() function.

Difference: In this there is no lines, that's why no frequency exist. whereas in contour plot frequency exits.


```python
from mpl_toolkits import mplot3d   # ax.plot_surface() Fn.
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X ** 2 + Y ** 2))

ax = plt.axes(projection='3d')    # Compulsary line for 3D plotation
ax. plot_surface(X, Y, Z, color='green')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('sin(circle) 3D Surface')
plt.show()
```
