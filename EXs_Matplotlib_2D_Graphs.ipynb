# 1) Line Chart :-

A line chart or line graph displays the evolution of
one or several numeric variables. Data points are
connected by straight line segments. A line chart is
often used to visualize a trend in data over
intervals of time. You can use the
plot(x,y) method to create a line chart.

```python
# import matplotlib.pyplot as plt 
# or we can use below method for importing pyplot module (which is also a short module of matplotlib module (big) )
from matplotlib import pyplot as plt
# .plot(), .show() Methods of matplotlib.pyplot module  
x=[1,2,3,4]
y=[12,22,7,34]
plt.plot(x,y)
plt.show()
```

```python
# You can also add the label on X-axis and Y-axis, and title to your chart.
# same as above example
from matplotlib import pyplot as plt
# .xlabel(), .ylable(), .title() Methods of matplotlib.pyplot module
x=[1,2,3,4]
y=[2,10,6,14]
plt.plot(x,y,color='green')
plt.xlabel('time')
plt.ylabel('velocity')
plt.title('My first Time-Velocity LINE CHART')
plt.show()
# here Data points are (1,2), (2,10), (3,6), (4,14) .
```

# 2) Scatter Plot :-

A Scatterplot displays the value of 2 sets of data on
2 dimensions. Each dot represents an observation.
The position on the X (horizontal) and Y (vertical)
axis represents the values of the 2 variables. It is
really useful to study the relationship between
both variables.
Use the scatter() method to draw a scatter plot
diagram.

Difference: In scatter plot we plot only Data points in the plot, whereas in Line chart we joined all the plotted points.

```python
# .scatter Method
from matplotlib import pyplot as plt
x=[10,8,20,16]
y=[5,15,2,11]
plt.scatter(x,y,color='red')
plt.xlabel('age')
plt.ylabel('class')
plt.title('My first SCATTR PLOT')
plt.show()
```

# 3) Bar Chart / Plot :-

Bar charts are used to
present categorical data with rectangular bars. The
bars can be plotted vertically or horizontally, and
their heights/lengths are proportional to the values
that they represent. Use the bar() method to draw
a bar plot diagram.


```python
# Verticale bar chart
# bar() Method
from matplotlib import pyplot as plt
subject=['maths','chemestry','physics','english']
marks=[99,95,90,85]
plt.xlabel('subject')
plt.ylabel('marks')
plt.title('My first BAR CHART / PLOT')
plt.bar(subject,marks,color='g')
plt.show()
```

```python
# Horizontale bar chart with same example
# .barh() Method
from matplotlib import pyplot as plt
subject=['maths','chemestry','physics','english']
marks=[99,95,90,85]
plt.xlabel('marks')
plt.ylabel('subject')
plt.title('My first BAR CHART / PLOT')
plt.barh(subject,marks,color='r')
plt.show()
```

# > Comparing Two Bar Charts:

```python
# .legend() Method is used for showing the labeled box in comparison chart
# we will take example for comparison between 1 to 10 odd vs even numbers
from matplotlib import pyplot as plt

# for odd numbers:
plt.bar([1,3,5,7,9],[5,2,7,8,2], label='Odd ones', color='g')
# for even numbers:
plt.bar([2,4,6,8,10],[8,6,2,5,6], label='Even ones', color='r')

# height of '1' is 5, whereas height of '2' is 8 . 
plt.legend()
plt.xlabel('Bar number')
plt.ylabel('Bar height')
plt.title('Odd numbers V/S Even nubers')
plt.show()
```

# 4) Histogram :-

It is a type of bar plot where X-axis represents the bin ranges while Y-axis gives information about frequency. Use the hist() method to draw a histogram diagram

```python
# .hist() Method
# Bins are just like groups
# here we made bins for obtaining marks in ranges 0-25, 25-50, 50-75, 75-100 .
import numpy as np
import matplotlib.pyplot as plt
arr=np.array([22,87,5,43,56,73,55,54,11,20,51,75,31,27])
plt.hist(arr, bins = [0,25,50,75,100], color='g')
plt.xlabel('Bins of optained marks')
plt.ylabel('Frequencies Of bins')
plt.title('My first HISTOGRAM')
plt.show()
# Note: 75 will be counted in range 75-100 .
```

```python
# continue...
# here we will take 20 random integers (repeat also possible) from range 1 to 100.
# here total no. of bins will be 10 as assigned below, which will be dividede acoording to given range.
x=np.random.randint(0,100,20)
print(x)
plt.hist(x, bins=5)
plt.show()
```

# 5) Area Plot/Chart :-

An area chart is really similar to a line chart, except
that the area between the x axis and the line is
filled in with color or shading. It represents the
evolution of a numerical variable following
another numerical variable. Use the
fill_between () method to draw a Area chart.

```python
# .fill_between Method
from matplotlib import pyplot as plt
x=[5,6,8,11]
y=[5,20,10,35]
#plt.plot(x,y) for cheking the line

plt.fill_between(x,y)
plt.title("My first AREA CHART")
plt.show()
# .xlabel, .ylabel can also used as before.
```

# 6) Pie Chart :-

A Pie Chart is a circular statistical plot that can
display only one series of data. Pie charts show the
size of items (called wedge) in one data series,
proportional to the sum of the items. The data
points in a pie chart are shown as a percentage of
the whole pie. The area of the wedge is
determined by the length of the arc of the wedge.
Use the pie () method to draw a Pie chart.


```python
# .pie() Method
from matplotlib import pyplot as plt
cars = ['AUDI','BMW','FORD','TESLA', 'JAGUAR', 'MERCEDES']
data = [25, 12, 38, 38, 12, 25]
plt.pie(data, labels = cars)
plt.title("My first PIE CHART")
plt.show()
# we can also use colors of our choice like:
#plt.pie(data, labels = cars, colors=['red','blue','green','yellow','black','orange'])
```

# 7) Boxplot :-

Boxplot is probably one of the most common type
of graphic. It gives a nice summary of one or
several numeric variables. The line that divides the
box into 2 parts represents the median of the data.
The end of the box shows the upper(1st) and
lower(3rd) quartiles. The extreme lines shows the
highest and lowest value excluding outliers.
Use the boxplot() method to draw a box plot.

```python
# .boxplot() Method
from matplotlib import pyplot as plt

data1= [100,120,140,160,180]
plt.boxplot(data1)
plt.title("My first BOXPLOT")
plt.show()

data2=[1,7,6,9,5,15,16,32,20,8,30]
plt.boxplot(data2)
plt.title("My second BOXPLOT")
plt.show()

data3=[1,7,6,9,5]
plt.boxplot(data3)
plt.title("My third BOXPLOT")
plt.show()
```

# 8) Contour Plot :-

 These contours are sometimes called
the z-slices. A contour plot is appropriate if you
want to see how value Z changes as a function of
two inputs X and Y, such that Z = f(X,Y) .
The independent variables x and y are usually
restricted to a regular grid called meshgrid. The
numpy.meshgrid creates a rectangular grid out of
an array of x values and an array of y values.Matplotlib API contains contour() and contourf()
functions that draw contour lines and filled
contours, respectively. Both functions need three
parameters x,y and z.

In short, The matplotlib.pyplot.contour() are usually useful
when Z = f(X, Y) i.e Z changes as a function of input
X and Y. A contourf() is also available which allows
us to draw filled contours.

```python
# .meshgrid() and .sqrt() of numpy
import numpy as np
# .contour(), .contourf() and .colorbar() of matplotlib.pyplot
import matplotlib.pyplot as plt
#Generating 100 numbers from -3.0 to 3.0
xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(xlist, ylist)

Z = np.sqrt(X**2 + Y**2)    # it is the Circle's equation
#cp = plt.contour(X, Y, Z)  # it gives only circle lines.
cp = plt.contourf(X, Y, Z)
plt.colorbar(cp) # Add a colorbar to a plot
plt.title("My first Contour Plot")
plt.show()

Z = np.sqrt(X**2/1 + Y**2/4)    # it is the elipse's equation
#cp = plt.contour(X, Y, Z)  # it gives only circle lines.
cp = plt.contourf(X, Y, Z, 20) # 20 is frequeny And we can also use cmap . see 3D exs.
plt.colorbar(cp) # Add a colorbar to a plot
plt.title("My second Contour Plot")
plt.show()
```
